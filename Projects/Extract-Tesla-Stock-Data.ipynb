### Question2

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0 Safari/537.36"
}
```

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
page = requests.get(url,headers=headers).text
```

```python
soup = BeautifulSoup(page,"html.parser")
soup
```

```python
revenue_table = soup.find('table',{"class":"historical_data_table table"})
revenue_table
```

```python
revenue_data = pd.DataFrame(columns=["Year","Annual Revenue(Millions of US $)"])
```

```python
rows = revenue_table.find_all('tr')
```

```python
for row in rows:
    cols = row.find_all('td')
    if len(cols) >= 2:
        year = cols[0].text
        annual_revenue = cols[1].text

        revenue_data = pd.concat([revenue_data,pd.DataFrame({"Year":[year],"Annual Revenue(Millions of US $)":[annual_revenue]})],ignore_index=True)

revenue_data
         
```
