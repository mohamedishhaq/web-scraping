```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "http://quotes.toscrape.com/"
page = requests.get(URL).text
```

```python
soup = BeautifulSoup(page, "html.parser")
soup
```

```python
quotes = soup.find_all('div',{"class":"quote"})
quotes
```

```python
quotes_data = pd.DataFrame(columns=["Quote","Autor"])
for quote in quotes:
    text = quote.find("span",{"class":'text'})
    autor = quote.find("small",{"class":'author'}).text

    quotes_data = pd.concat([quotes_data,pd.DataFrame({'Quote':[text],"Autor":[autor]})],ignore_index=True)
    
quotes_data
    
    
```
